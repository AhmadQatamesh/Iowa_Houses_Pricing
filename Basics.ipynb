```python
def transform_features(df):
    data = df.copy()
    
    cols_mv = data.isnull().sum()/len(data)*100
    data = df[cols_mv[cols_mv<5].index]
    
    text_cols_mv = data.select_dtypes(include='object').isnull().sum()
    text_cols_mv_tags = text_cols_mv[text_cols_mv>0].index
    data = data.drop(text_cols_mv_tags,axis=1)
    
    num_mv = data.select_dtypes(include = ['float','int']).isnull().sum()
    num_mv_tags = num_mv[num_mv>0].index
    replacement_dict = data[num_mv_tags].mode().to_dict(orient='records')[0]
    data = data.fillna(replacement_dict)
    
    years_remod = data['Yr Sold']-data['Year Remod/Add']
    data = data.drop([1702,2180,2181],axis=0)
    data['years_remod'] = years_remod
    data['years_sold'] = years_sold
    data = data.drop(["Year Built", "Year Remod/Add"], axis = 1)
    data = data.drop(["Mo Sold", "Sale Condition", "Sale Type", "Yr Sold"], axis=1)
    data = data.drop(["PID", "Order"], axis = 1)
    
    return data

def select_features(df,cutoff=0.4,nr_category=10):
    corr = df.corr()['SalePrice'].abs().sort_values(ascending=False)
    low_corr_cols = corr[(corr<cutoff)].index
    data = df.drop(low_corr_cols,axis=1)
    data = data.drop('Garage Area',1)
    
    nominal_features = ["PID", "MS SubClass", "MS Zoning", "Street", "Alley", "Land Contour", "Lot Config", "Neighborhood", 
                    "Condition 1", "Condition 2", "Bldg Type", "House Style", "Roof Style", "Roof Matl", "Exterior 1st", 
                    "Exterior 2nd", "Mas Vnr Type", "Foundation", "Heating", "Central Air", "Garage Type", 
                    "Misc Feature", "Sale Type", "Sale Condition"]
    transform_cat_cols = []
    for col in nominal_features:
        if col in data.columns:
            transform_cat_cols.append(col)
            
    uniqueness_counts = data[transform_cat_cols].apply(lambda col: len(col.value_counts())).sort_values()
    drop_nonuniq_cols = uniqueness_counts[uniqueness_counts > nr_category].index
    data = data.drop(drop_nonuniq_cols, axis=1)
    
    text_cols = data.select_dtypes(include=['object'])
    for col in text_cols:
        data[col] = data[col].astype('category')
    data=pd.concat([data,pd.get_dummies(data.select_dtypes(include='category'))],axis=1)
    return data

def train_test(df):
    train = df[:1460]
    test = df[1460:]
    numerical_cols=df.select_dtypes(include=['float','integer']).columns.drop('SalePrice')
    lr = LinearRegression()
    lr.fit(train[numerical_cols],train['SalePrice'])
    prediction = lr.predict(test[numerical_cols])
    coef = lr.coef_
    intercept = lr.intercept_
    test_mse = mean_squared_error(y_true = test['SalePrice'],y_pred = prediction)
    test_rmse = np.sqrt(test_mse)
    return(test_rmse)

data=pd.read_csv('AmesHousing.tsv','\t')
transform_df = transform_features(data)
filtered_df = select_features(transform_df,0.3,5)
rmse = train_test(filtered_df)
rmse
```

Next iteration I will use KFold to test diffrent df splits and shuffle so I can minimize the error. This is a holdout validation that will give me a better idea of how I can optimize the number of folds

I will add k variable to train_test function. k=0 will return the same previous logic, k=1 will shuffle and return average rmse for prediction on both test and train splits. k>1 will use KFold to return average rmse for #k of splits

```python
def train_test(df,k=0):
    numerical_cols = df.select_dtypes(include=['float','integer']).columns.drop('SalePrice')
    lr = LinearRegression()
    if k==0:
        train = df[:1460]
        test = df[1460:]
        lr.fit(train[numerical_cols],train['SalePrice'])
        prediction = lr.predict(test[numerical_cols])
        coef = lr.coef_
        intercept = lr.intercept_
        test_mse = mean_squared_error(y_true = test['SalePrice'],y_pred = prediction)
        test_rmse = np.sqrt(test_mse)
        return test_rmse
    if k==1:
        # Randomize *all* rows (frac=1) from `df` and return
        shuffled_df = df.sample(frac=1, )
        train = shuffled_df[:1460]
        test = shuffled_df[1460:]
        
        lr.fit(train[numerical_cols], train["SalePrice"])
        predictions_one = lr.predict(test[numerical_cols])        
        
        mse_one = mean_squared_error(test["SalePrice"], predictions_one)
        rmse_one = np.sqrt(mse_one)
        
        lr.fit(test[numerical_cols], test["SalePrice"])
        predictions_two = lr.predict(train[numerical_cols])        
       
        mse_two = mean_squared_error(train["SalePrice"], predictions_two)
        rmse_two = np.sqrt(mse_two)
        
        avg_rmse = np.mean([rmse_one, rmse_two])
        print(rmse_one)
        print(rmse_two)
        return avg_rmse
    else:
        kf = KFold(n_splits=k,shuffle=True,random_state=None)
        rmse = []
        for train_index, test_index in kf.split(df):
            train = df.iloc[train_index]
            test = df.iloc[test_index]
            lr.fit(train[numerical_cols],train['SalePrice'])
            prediction = lr.predict(test[numerical_cols])
            test_prediction = lr.predict(test[numerical_cols])
            # train_prediction = lr.predict(train[numerical_cols])
            test_rmse = np.sqrt(mean_squared_error(test['SalePrice'],test_prediction))
            # train_rmse = np.sqrt(mean_squared_error(train['SalePrice'],train_preidciton))
            rmse.append(test_rmse)
        print(rmse)
        avg_rmse = np.mean(rmse)
        return(avg_rmse)

data = pd.read_csv('AmesHousing.tsv','\t')
transform_df = transform_features(data)
filtered_df = select_features(transform_df,0.3,5)
rmse = train_test(filtered_df,5)
rmse
```
